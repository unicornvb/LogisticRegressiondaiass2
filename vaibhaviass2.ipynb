```python
#separating features and target
x=data.drop(columns=["y"])
y=data.y
```

```python
# Convert 'yes'/'no' in y to numerical values
y = y.map({'yes': 1, 'no': 0})
```

```python
y
```

```python
# splits data into random train and validation (or test) subsets
from sklearn.model_selection import train_test_split
x_train, x_valid, y_train, y_valid = train_test_split(x,y,train_size=0.8)
```

```python
# converting the categorical column (housing )to numerical by binary encoding 
import category_encoders as ce

encoder = ce.BinaryEncoder(cols=['housing'])
x_train= encoder.fit_transform(x_train)
x_valid = encoder.transform(x_valid)
```

```python
#deleting the extra column generated in the process
x_train.drop(columns=["housing_0"],inplace=True)
x_valid.drop(columns=["housing_0"],inplace=True)
```

```python
# converting the categorical column ( loan )to numerical by binary encoding
encoder = ce.BinaryEncoder(cols=['loan'])
x_train= encoder.fit_transform(x_train)
x_valid= encoder.transform(x_valid)
#deleting the extra column generated in the process
x_train.drop(columns=["loan_0"],inplace=True)
x_valid.drop(columns=["loan_0"],inplace=True)
```

```python
# converting the categorical column (month)to numerical by label encoding
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
x_train['Month'] = le.fit_transform(x_train['month'])
x_valid['Month'] = le.transform(x_valid['month'])
```

```python
#deleting the old column
x_train.drop(columns=["month"],inplace=True)
x_valid.drop(columns=["month"],inplace=True)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Initializing the SVM model
svm_model = SVC(kernel='linear', random_state=42) 

# Train the model
svm_model.fit(x_train, y_train)

# Make predictions on the test set
y_pred = svm_model.predict(x_valid)

# Evaluating the model
accuracy = accuracy_score(y_valid, y_pred)
print(f'Accuracy: {accuracy:.2f}')

# Printing classification report
print("\nClassification Report:")
print(classification_report(y_valid, y_pred))

# Printing confusion matrix
print("\nConfusion Matrix:")
print(confusion_matrix(y_valid, y_pred))
```
